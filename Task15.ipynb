# **Removing Outliers**

**1. Z-score Method**

```python
upper_limit = ddf['bath'].mean()+3*ddf['bath'].std()
lower_limit = ddf['bath'].mean()-3*ddf['bath'].std()
```

```python
upper_limit
```

```python
lower_limit
```

```python
# Finding the outliers
ddf.loc[(ddf['bath'] > upper_limit) | (ddf['bath'] < lower_limit)]
```

```python
# trimming - delete the outlier data
new_df = ddf.loc[(ddf['bath'] <= upper_limit) | (ddf['bath'] >= lower_limit)]
print("before removing outliers : ",len(ddf))
print("after removing outliers : ",len(new_df))
print("outliers :" ,len(ddf)-len(new_df))
```

```python
sns.boxplot(new_df['bath'])
```

```python
# capping - change the outlier values to upper or lower limit values
new_df = ddf.copy()
new_df.loc[new_df['bath']>upper_limit,'bath'] = upper_limit
new_df.loc[new_df['bath']<lower_limit,'bath'] = lower_limit
```

```python
sns.boxplot(new_df['bath'])
```

```python
len(new_df)
```

**2. IQR(Inter Quartile Range) Method**

```python
q1 = ddf['bath'].quantile(0.25)
q3 = ddf['bath'].quantile(0.75)
iqr = q3 - q1
```

```python
q1,q3,iqr
```

```python
upper_limit = q3 +(1.5*iqr)
lower_limit = q1 -(1.5*iqr)
```

```python
upper_limit , lower_limit
```

```python
sns.boxplot(ddf['bath'])
```

```python
# Finding the outliers
ddf.loc[(ddf['bath'] > upper_limit) | (ddf['bath'] < lower_limit)]
```

```python
# trimming - delete the outlier data
new_df = ddf.loc[(ddf['bath'] <= upper_limit) | (ddf['bath'] >= lower_limit)]
print("before removing outliers : ",len(ddf))
print("after removing outliers : ",len(new_df))
print("outliers :" ,len(ddf)-len(new_df))
```

```python
sns.boxplot(new_df['bath'])
```

```python
# capping - change the outlier values to upper or lower limit values
new_df = ddf.copy()
new_df.loc[new_df['bath']>upper_limit,'bath'] = upper_limit
new_df.loc[new_df['bath']<lower_limit,'bath'] = lower_limit
```
