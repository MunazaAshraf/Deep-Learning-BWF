```python
import numpy as np
import pandas as pd
```

**LOADING ALL FILES AT ONCE USING GLOB AND CONCATTING ALL FILES USING CONCAT()**

```python
import glob
import os
```

```python
all_files = glob.glob("/content/US_Census/*.csv")
```

```python
us_census = pd.concat((pd.read_csv(f) for f in all_files))
us_census
```

```python
us_census['Hispanic'] = us_census['Hispanic'].apply(lambda x: (float(str(x).replace('%', ''))))
us_census['White'] = us_census['White'].apply(lambda x: (float(str(x).replace('%', ''))))
us_census['Black'] = us_census['Black'].apply(lambda x: (float(str(x).replace('%', ''))))
us_census['Native'] = us_census['Native'].apply(lambda x: (float(str(x).replace('%', ''))))
us_census['Asian'] = us_census['Asian'].apply(lambda x: (float(str(x).replace('%', ''))))
us_census['Pacific'] = us_census['Pacific'].apply(lambda x: (float(str(x).replace('%', ''))))


us_census
```

```python
us_census[['Male', 'Female']] = us_census.GenderPop.str.split("_", expand = True)
us_census
```

```python
us_census['Male'] = us_census['Male'].replace('M', '', regex=True)
us_census['Female'] = us_census['Female'].replace('F', '', regex=True)
```

```python
us_census
```

```python
us_census = us_census.drop('GenderPop', axis=1)
us_census
```

```python
us_census['Female']=us_census['Female'].replace('', np.nan)
us_census
```

```python
us_census['Male'] = us_census['Male'].astype(int)
```

```python
# filling null values in female column with TotalPop of that state minus the Men for that state.
nanvalue = us_census['TotalPop'] - us_census['Male']
```

```python
us_census['Female'].fillna(nanvalue, inplace=True)
```

```python
us_census
```

```python
us_census.Female.isnull().sum()
```

```python
us_census['Female'] = us_census['Female'].astype(int)
```

```python
us_census.Female.dtypes
```

```python
us_census['Income'] = us_census['Income'].str.replace('$', '')
us_census
```

```python
us_census['Income'] = us_census['Income'].astype(float)
```
