# **Combining Datasets**

**Using Join()**

```python
import numpy as np
import pandas as pd
```

```python
# one-to-one joins
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})

df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
                    'hire_date': [2004, 2008, 2012, 2014]})
```

```python
df1
```

```python
df2
```

```python
df3 = pd.merge(df1, df2)
df3
```

```python
# many-to-one joins
df4 = pd.DataFrame({'group': ['Accounting', 'Engineering', 'HR'],
                    'supervisor': ['Carly', 'Guido', 'Steve']})
df4
```

```python
df5 = pd.merge(df3, df4)
df5
```

```python
# many-to-many joins
df5 = pd.DataFrame({'group': ['Accounting', 'Accounting',
                              'Engineering', 'Engineering', 'HR', 'HR'],
                    'skills': ['math', 'spreadsheets', 'coding', 'linux',
                               'spreadsheets', 'organization']})
df5
```

```python
df6 = pd.merge(df1,df5)
df6
```

**The on keyword**

```python
pd.merge(df1, df2, on='employee')
```

**The left_on and right_on keywords**

```python
dff = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'salary': [70000, 80000, 120000, 90000]})
dff
```

```python
pd.merge(df1, dff, left_on="employee", right_on="name")
```

```python
pd.merge(df1, dff, left_on="employee", right_on="name").drop('name', axis=1)
```

**The left_index and right_index keywords**

```python
df1a = df1.set_index('employee')
df2a = df2.set_index('employee')
```

```python
df1a
```

```python
df2a
```

```python
pd.merge(df1a, df2a, left_index=True, right_index=True)
```

**Using join method**

```python
df1a.join(df2a)
```

```python
pd.merge(df1a, dff, left_index=True, right_on='name')
```

**Specifying Set Arithmetic for Joins**

```python
df6 = pd.DataFrame({'name': ['Peter', 'Paul', 'Mary'],
                    'food': ['fish', 'beans', 'bread']},
                   columns=['name', 'food'])
df6
```

```python
df7 = pd.DataFrame({'name': ['Mary', 'Joseph'],
                    'drink': ['wine', 'beer']},
                   columns=['name', 'drink'])
df7
```

```python
pd.merge(df6, df7)
```

```python
pd.merge(df6, df7, how='inner')
```

```python
pd.merge(df6, df7, how='outer')
```

```python
pd.merge(df6, df7, how='left')
```

```python
pd.merge(df6, df7, how='right')
```

**Overlapping Column Names: The suffixes Keyword**

```python
df8 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'rank': [1, 2, 3, 4]})
df8
```

```python
df9 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'rank': [3, 1, 4, 2]})
df9
```

```python
pd.merge(df8, df9, on="name")
```

```python
pd.merge(df8, df9, on="name", suffixes=["_L", "_R"])
```

# **Applying on Datasets**

```python
abv = pd.read_excel('/content/file1.xlsx')
abv
```

```python
areas = pd.read_excel('/content/file2.xlsx')
areas
```

```python
pop = pd.read_excel('/content/file3.xlsx')
pop
```

```python
merged = pd.merge(pop, abv, how='outer', left_on='state/region', right_on='abbreviation')
merged
```

```python
merged.duplicated() # to check for duplicate data
```

```python
merged.isnull()
```

```python
merged.isnull().sum()
```

```python
merged = merged.drop('abbreviation', 1) # drop duplicate info
merged.head()
```

```python
merged.isnull().any()
```

```python
merged[merged['population'].isnull()].head()
```

```python
merged.loc[merged['state'].isnull(), 'state/region'].unique()
```

```python
merged.loc[merged['state/region'] == 'PR', 'state'] = 'Puerto Rico'
merged.loc[merged['state/region'] == 'USA', 'state'] = 'United States'
merged.isnull().any()
```

```python
final = pd.merge(merged, areas, on='state', how='left')
final.head()
```

```python
final.isnull().any()
```

```python
final['state'][final['area (sq. mi)'].isnull()].unique()
```

```python
final.dropna(inplace=True)
final.head()
```

```python
data2010 = final.query("year == 2010 & ages == 'total'")
data2010.head()
```

```python
data2010.set_index('state', inplace=True)
density = data2010['population'] / data2010['area (sq. mi)']
```

```python
density.sort_values(ascending=False, inplace=True)
density.head()
```

```python
density.tail()
```
